```python
import openai
import os

openai.api_key = os.environ.get('OPENAI_API_KEY')

def sent_gpt_request(topic, answer, retries=3):
    print("正在请求...")

    retries_total = 0
    while retries_total < retries:
        try:
            completion = openai.ChatCompletion.create(
                model="gpt-3.5-turbo-16k-0613",
                messages=[
                    {"role": "user", "content": """你的任务是把一些试题与答案匹配。"""},
                    {"role": "assistant", "content": "请完善你的需求。"},
                    {"role": "user", "content": """我会以list结构向你提供试题，会以字符串的格式向你提供答案，你不需要处理我提供的json中的数据，你需要把"answer"字段当成答案即可，这些题目都是选择题。我需要的格式如下：\n [{'topic': '1．函数 的定义域为 （ ）', "answer":"A"}, ...]。。"""},
                    {"role": "assistant", "content": "我明白了你的需求。现在，请提供你的试卷。"},
                    {"role": "user", "content":topic},
                    {"role": "assistant", "content": "请提供你的答案。"},
                    {"role": "user", "content":answer},
                    {"role": "assistant", "content": '这是我找到的选择题的JSON数据:\n'}
                ],
                temperature=0.1,
            )
            return completion.choices[0]["message"]["content"]
        except Exception as e:
            retries_total += 1
            print(f"请求失败重新请求中... {e}")

    return None
```
